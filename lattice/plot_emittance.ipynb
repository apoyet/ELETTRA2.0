```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
from cpymad.madx import Madx
from matplotlib import cm, gridspec, patches

%matplotlib inline
%config InlineBackend.figure_format = "retina"

import elettra_toolbox

import seaborn as sns
```

```python
emit_df = pd.read_parquet('emit_df_vader.parquet')
```

```python
k_vader_0 = -1.962662

k_plot = np.array(emit_df.index) - k_vader_0
```

```python
plt.figure(figsize=(10,7))
plt.plot(k_plot*1e2, emit_df['ex'].to_numpy()*1e12, '-', lw=2)
plt.xlabel('$\\Delta k_{vader}$ [x 10$^{-2}$]', fontsize=16)
plt.ylabel('$\\varepsilon_x$ [pm]', fontsize=16)
plt.tick_params(labelsize=14)
```

## Quad scan with optimized AB
---

```python
emit_df = pd.read_pickle('emit_df_quad_opt.pkl')
```

```python
emit_df_plot = emit_df.copy()

k1_qd1_0          =       -3.420000
k1_qf1_0          =       5.735000

scaling_factor = 1
emit_df_plot.index = (emit_df_plot.index - k1_qd1_0)*scaling_factor
emit_df_plot.columns = (emit_df_plot.columns - k1_qf1_0)*scaling_factor
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.heatmap(emit_df_plot*1e12, cmap='RdBu', vmin=160, vmax=170)
ax.set_yticklabels(["{:.1f}".format(x) for x in [float(ax.get_yticklabels()[i].get_text())*1e2 for i in range(len(ax.get_yticklabels()))]])
ax.set_xticklabels(["{:.1f}".format(x) for x in [float(ax.get_xticklabels()[i].get_text())*1e2 for i in range(len(ax.get_xticklabels()))]]);
```
